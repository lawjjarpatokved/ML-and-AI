## Descriptive Statistics

Descriptive statistics involve a set of summary measures that provide a snapshot of the dataset's characteristics. These measures help us understand the distribution, central tendency, and variability within the data.

- Mean: The average value of the data.
- Median: The middle value when the data is sorted.
- Mode: The most frequently occurring value.
- Range: The difference between the maximum and minimum values.
- Standard Deviation: A more interpretable measure of data spread.
These statistics provide a preliminary understanding of the dataset, which is valuable for subsequent analysis and decision-making.

# Bivariate Analysis


```python
from scipy.stats import chi2_contingency

addr_n_severitycode_df=pd.crosstab(df['ADDRTYPE'],
                                           df["SEVERITYCODE"]
                                           )
addr_n_severitycode_df
```

# Most accidents that occur at blocks are of severitycode 1

```python
chi2, p, dof, expected= chi2_contingency(addr_n_severitycode_df)
p
```

# P value shows that there is  strong relation between addrtype and severitycode. Now trying other features

```python
weath_light_fatalities_df = pd.crosstab([df.LIGHTCOND, df.WEATHER], 
								df.FATALITIES, margins = True) 

chi2, p, dof, expected= chi2_contingency(weath_light_fatalities_df)
p
```

```python
"""KOLMOGOROV-SMIRNOV TESTS"""

from scipy.stats import kstest, shapiro

statistic, pvalue = kstest(df['SERIOUSINJURIES'], 'norm')
print("p-value",pvalue)
```

# this shows that no of people seriously injured does not follow normal distribution

```python
statistic, pvalue = kstest(df['FATALITIES'], 'norm')
print("p-value",pvalue)

statistic, pvalue = kstest(df['INJURIES'], 'norm')
print("p-value",pvalue)
```

```python
"""LIGHT CONDITION AND WEATHER CONDITION VS ACCIDENT SEVERITY"""

x=pd.crosstab([df.LIGHTCOND],df["SEVERITYDESC"])
y=pd.crosstab([df.WEATHER],df["SEVERITYDESC"])

x.plot(kind="barh",stacked=True)
plt.title("LIGHT COND VS ACCIDENT SEVERITY")
plt.xlabel("NUMBER OF ACCIDENTS")
plt.show()

y.plot(kind="barh",stacked=True)
plt.title("WEATHER VS ACCIDENT SEVERITY")
plt.xlabel("NUMBER OF ACCIDENTS")
plt.show()
```

```python
"""LIGHT CONDITION AND WEATHER CONDITION VS ACCIDENT SEVERITY"""

new_df=df["FATALITIES"].fillna("N")
x=pd.crosstab(new_df,df["SEVERITYDESC"])

x.plot(kind="barh",stacked=True)
plt.title("FARALITIES VS ACCIDENT SEVERITY")
plt.xlabel("NUMBER OF ACCIDENTS")
plt.show()

x=pd.crosstab(df["COLLISIONTYPE"],df["SEVERITYDESC"])

x.plot(kind="barh",stacked=True)
plt.title("COLLISION TYPE VS ACCIDENT SEVERITY")
plt.xlabel("NUMBER OF ACCIDENTS")
plt.show()

x=pd.crosstab(df["JUNCTIONTYPE"],df["SEVERITYDESC"])

x.plot(kind="barh",stacked=True)
plt.title("LOCATION VS ACCIDENT SEVERITY")
plt.xlabel("NUMBER OF ACCIDENTS")
plt.show()


x=pd.crosstab(df["ADDRTYPE"],df["SEVERITYDESC"])

x.plot(kind="barh",stacked=True)
plt.title("ADDRESS VS ACCIDENT SEVERITY")
plt.xlabel("NUMBER OF ACCIDENTS")
plt.show()
```
